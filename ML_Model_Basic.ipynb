#**Load Data**
Pandas is a widely-used Python library in data science and machine learning for data manipulation, analysis, and cleaning.

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

#**Data Preparation**

##**Data Seperation as X and Y**


```python
y = df['logS']
y
```

*   Removing Y (output/dependent feature) from the data
*   axis=1 (Work in columns), axis=0 (Work in rows)





```python
x = df.drop('logS', axis=1)
x
```

##**Data Splitting**
Scikit-learn (also known as sklearn) is a popular open-source machine learning library for Python. It provides a wide range of tools and algorithms for various machine learning tasks


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

#**Model Building**

##**Linear Regression**
Linear regression is a supervised machine learning algorithm used for predicting a continuous target variable based on one or more input features. It models the relationship between the dependent variable (y) and independent variable(s) (x) by fitting a straight line through the data.

###Training the model by lr.fit()

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)
```

###Applying the model to make predictions

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_test_pred
```

###Evaluate Model
Compare predictions with actual data

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

##Random Forest
Random Forest is a supervised machine learning algorithm that can be used for both classification and regression tasks. It builds multiple decision trees during training and combines their outputs to make predictions. Random Forest is an ensemble method, meaning it aggregates the results of multiple models to improve accuracy and reduce overfitting.

###Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)

rf.fit(x_train, y_train)
```

###Applying the model to make predictions

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

###Evaluate Model

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

##**Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```
